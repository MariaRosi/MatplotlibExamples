```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data_for_fill_line_plot.csv')
```

```python
df.tail(5)
```

```python
ages = df['Age']
all_dev_sal = df['All_Devs']
py_dev_sal = df['Python']
js_dev_sal = df['JavaScript']
```

```python
plt.style.use('ggplot')
overall_median = 57287

plt.plot(ages, all_dev_sal, label='all_dev')
plt.plot(ages, py_dev_sal, label='python')

plt.fill_between(ages, py_dev_sal, overall_median, alpha=0.25)

plt.title('Salaries Per Programming Languages On Age Ranges', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
plt.style.use('ggplot')
overall_median = 57287

plt.plot(ages, all_dev_sal, label='all_dev')
plt.plot(ages, py_dev_sal, label='python')

plt.fill_between(ages,
                 py_dev_sal,
                 overall_median,
                 where=(py_dev_sal > overall_median),
                 color='g',
                 interpolate=True,
                 alpha=0.25
                )

plt.fill_between(ages,
                 py_dev_sal,
                 overall_median,
                 where=(py_dev_sal <= overall_median),
                 color='r',
                 interpolate=True,
                 alpha=0.25
                )


plt.title('Salaries Per Programming Languages On Age Ranges', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.legend()
plt.tight_layout()

plt.show()
```
