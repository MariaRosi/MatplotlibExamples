```python
import matplotlib.pyplot as plt
```

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
slice_labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.style.use('fivethirtyeight')

plt.pie(slices,
        labels=slice_labels,
        wedgeprops={'edgecolor': 'black'}
       )

plt.title('Language Popularity Pie Chart')
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
slice_labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slice_explode = [0, 0, 0, 0.2, 0]


plt.style.use('fivethirtyeight')

plt.pie(slices,
        labels=slice_labels,
        explode=slice_explode,
        shadow=True,
        startangle=90,
        wedgeprops={'edgecolor': 'black'}
       )

plt.title('Language Popularity Pie Chart')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.pie(slices,
        labels=slice_labels,
        explode=slice_explode,
        shadow=True,
        startangle=90,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'}
       )

plt.title('Language Popularity Pie Chart')
plt.tight_layout()
plt.show()
```
