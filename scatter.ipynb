```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_for_scatter.csv')
```

```python
df
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.style.use('fivethirtyeight')

plt.scatter(view_count,
            likes,
            c=ratio,
            cmap='summer',
            edgecolor='k',
            alpha=0.75
           )

plt.xscale('log')
plt.yscale('log')

bar = plt.colorbar()
bar.set_label('Like/Dislike Ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```
