```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_for_subplot.csv')
df.head()
```

```python
ages = df['Age']
all_dev = df['All_Devs']
py_dev = df['Python']
js_dev = df['JavaScript']
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(ages,
         all_dev,
         color='g',
         alpha=0.50,
         marker='o',
         markersize=2,
         linestyle='--',
         linewidth=2,
         label='other_dev'
        )
plt.title('Salary For All Other Developer', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)

plt.subplot(2,2,2)
plt.plot(ages,
         py_dev,
         color='r',
         alpha=0.50,
         marker='o',
         markersize=2,
         linestyle='--',
         linewidth=2,
         label='py_dev'
        )
plt.title('Salary For Python Developer', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)

plt.subplot(2,2,3)
plt.plot(ages,
         js_dev,
         color='y',
         alpha=0.50,
         marker='o',
         markersize=2,
         linestyle='--',
         linewidth=2,
         label='js_dev'
        )
plt.title('Salary For JavaScript Developer', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.tight_layout(pad=4.0)

plt.show()
```
